```python
import feedparser
from urllib.parse import quote

# Consulta Google News só com site Inovação Tecnológica
query = "site:inovacaotecnologica.com.br"
url = f"https://news.google.com/rss/search?q={quote(query)}&hl=pt-BR&gl=BR&ceid=BR:pt-419"

# Lê o feed
feed = feedparser.parse(url)

# Exibe as notícias
for entry in feed.entries:
    print(f"Título: {entry.title}")
    print(f"Publicado em: {entry.published}")
    print(f"Link: {entry.link}")
    print("—" * 60)
```
