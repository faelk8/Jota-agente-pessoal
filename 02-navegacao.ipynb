```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url_base = 'https://www.inovacaotecnologica.com.br/noticias/assuntos.php?assunto=eletronica'

# Faz a requisição da página
response = requests.get(url_base)
soup = BeautifulSoup(response.text, 'html.parser')

# Extrai todos os links (tags <a href>)
urls = set()  # usar set para evitar duplicadas

for link in soup.find_all('a', href=True):
    href = link['href']
    full_url = urljoin(url_base, href)  # resolve links relativos
    urls.add(full_url)

# Mostra todas as URLs encontradas
for u in sorted(urls):
    print(u)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url_base = 'https://www.inovacaotecnologica.com.br/'

# Requisição da página inicial
response = requests.get(url_base)
soup = BeautifulSoup(response.text, 'html.parser')

# Encontra o link "Mais notícias"
mais_noticias = soup.find('a', string='Mais notícias')

if mais_noticias:
    href = mais_noticias['href']
    next_page_url = urljoin(url_base, href)

    print("➡️ Link para a próxima página:", next_page_url)

    # Agora você pode fazer uma nova requisição para essa URL
    next_response = requests.get(next_page_url)
    next_soup = BeautifulSoup(next_response.text, 'html.parser')

    # A partir daqui, você pode continuar coletando mais URLs dessa nova página
    for link in next_soup.find_all('a', href=True):
        full_url = urljoin(next_page_url, link['href'])
        print(full_url)
else:
    print("❌ Link 'Mais notícias' não encontrado.")
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

base_url = 'https://www.inovacaotecnologica.com.br/noticias/assuntos.php?assunto=eletronica'
base_url_sem_base = base_url.split('&base=')[0]

all_links = set()
base = 0
incremento = 15

while True:
    if base == 0:
        url = base_url_sem_base
    else:
        url = f"{base_url_sem_base}&base={base}"

    print(f"Buscando links na página: {url}")
    resp = requests.get(url)
    if resp.status_code != 200:
        print("Erro na requisição ou página não encontrada.")
        break

    soup = BeautifulSoup(resp.text, 'html.parser')

    links = soup.find_all('a', href=True)
    
    # Resolve links relativos para absolutos e filtra os links que contêm '/noticia.php?artigo='
    links_validos = []
    for link in links:
        href = link['href']
        full_url = urljoin('https://www.inovacaotecnologica.com.br/noticias/', href)
        if '/noticia.php?artigo=' in full_url:
            links_validos.append(full_url)

    if not links_validos:
        print("Nenhum link válido encontrado nesta página, finalizando.")
        break

    before_count = len(all_links)
    all_links.update(links_validos)
    after_count = len(all_links)

    print(f"Links coletados nesta página: {len(links_validos)} | Total acumulado: {after_count}")

    if before_count == after_count:
        print("Nenhum link novo encontrado, finalizando a busca.")
        break

    base += incremento

print("\nTotal de links únicos coletados:", len(all_links))

for link in sorted(all_links):
    print(link)
```
